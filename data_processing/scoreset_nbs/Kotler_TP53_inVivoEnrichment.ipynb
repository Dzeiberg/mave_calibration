```python
remove = set(('Ins','Del','IARC.InDel.GeneVars','Del2AA','DelAA','STOP','Sub2bp','IARC.InDel.SomatMuts'))
```

```python
subset = data[~data.Mut_type.isin(remove)]
```

```python
data[data.Mut_type == "Sub2bp"]
```

```python
subset
```

```python
from Bio.PDB.Polypeptide import protein_letters_3to1
protein_letters_1to3 = {v: k.title() for k, v in protein_letters_3to1.items()}
protein_letters_1to3["*"] = "Ter"
def make_hgvs_pro(r):
    return f"p.{protein_letters_1to3[r.AA_change[0]]}{r.Codon_num}{protein_letters_1to3[r.AA_change[-1]]}"
```

```python
subset = subset.assign(hgvs_pro=subset.apply(make_hgvs_pro, axis=1))
subset = subset.rename(columns={'Enrichment_invivo':'score'})
```

```python
subset.to_csv(dataset_path / "scoreset.csv", index=False)
```

```python
import pandas as pd
from pathlib import Path

dataset_path = Path("/mnt/i/bio/mave_curation/Kotler_TP53_inVivoEnrichment/")
```

```python
data = pd.read_excel(dataset_path / "raw" / 'mmc6.xlsx')
```

```python
data
```

```python
data.Mut_type.unique()
```

```python
subset = data[(~data.AA_change.isna()) & (data.AA_change.str.len() == 3) & (data.Mut_type != "DelAA")]
```

```python
subset
```

```python
from Bio.PDB.Polypeptide import protein_letters_3to1
protein_letters_1to3 = {v: k.title() for k, v in protein_letters_3to1.items()}
protein_letters_1to3["*"] = "Ter"
def make_hgvs_pro(r):
    return f"{protein_letters_1to3[r.AA_change[0]]}{r.Position}{protein_letters_1to3[r.AA_change[-1]]}"
```

```python
subset = subset.assign(hgvs_pro=subset.apply(make_hgvs_pro, axis=1))
subset = subset.rename(columns={'Enrichment_invivo':'score'})
```

```python
subset
```

```python
subset.to_csv(dataset_path / "scoreset.csv", index=False)
```

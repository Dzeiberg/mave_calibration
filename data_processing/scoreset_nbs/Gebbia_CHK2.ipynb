```python
import pandas as pd
from pathlib import Path
import json
dataset_path = Path("/data/dzeiberg/mave_calibration/data/Gebbia_CHK2/")
assert dataset_path.exists()
```

```python
data = pd.read_csv(dataset_path / 'raw' / 'media-1.csv')
data = data.assign(hgvs_pro=data.hgvs_pro.str.replace("*",'Ter'))
data = data.rename(columns=dict(score='raw_score',
                                error='raw_score_error'))
data = data.rename(columns=dict(refined_score='score',
                                refined_score_se='score_se'))
```

```python
data.to_csv(dataset_path / 'scoreset.csv',index=False)
```

```python
import pandas as pd
from pathlib import Path
import json
dataset_path = Path("/mnt/i/bio/mave_curation/Gebbia_CHK2/")
assert dataset_path.exists()
```

```python
data = pd.read_csv(dataset_path / 'raw' / 'media-1.csv')
```

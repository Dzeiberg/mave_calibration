```python

import pandas as pd
from pathlib import Path
import json
dataset_path = Path("/mnt/i/bio/mave_curation/Radford_DDX3X_survivalAssessmentAssay/")
assert dataset_path.exists()

data = pd.read_excel(dataset_path / "raw" / "Supplementary_Data_S14_SGE_data_for_all_variants_tested_with_annotation.xlsx")
```

```python
coding_subset = data.dropna(subset=['ref_aa','alt_aa'])
```

```python
coding_subset
```

```python
from Bio.PDB.Polypeptide import protein_letters_3to1
protein_letters_1to3 = {v: k.title() for k, v in protein_letters_3to1.items()}
protein_letters_1to3['STOP'] = "Ter"
def make_hgvs_pro(r):
    return f"p.{protein_letters_1to3[r.ref_aa]}{r.Protein_position}{protein_letters_1to3[r.alt_aa]}"
```

```python
coding_subset = coding_subset.assign(hgvs_pro=coding_subset.apply(make_hgvs_pro,axis=1))
```

```python
coding_subset = coding_subset.rename(columns={"D15_combined_LFC":"score"})
```

```python
coding_subset.loc[:,['hgvs_pro','score']]
```

```python
coding_subset[coding_subset.hgvs_pro.str[2:5] == coding_subset.hgvs_pro.str[-3:]]
```

```python
scoreset = pd.DataFrame(coding_subset.groupby('hgvs_pro').score.mean()).reset_index()
```

```python
scoreset
```

```python
scoreset.to_csv(dataset_path / 'scoreset.csv',index=False)
```

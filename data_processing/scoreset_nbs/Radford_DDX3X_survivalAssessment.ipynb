```python

import pandas as pd
pd.set_option('display.max_columns', None)
from pathlib import Path
import json
dataset_path = Path("/data/dzeiberg/mave_calibration/data/Radford_DDX3X_survivalAssessmentAssay/")
assert dataset_path.exists()

data = pd.read_excel(dataset_path / "raw" / "Supplementary_Data_S14_SGE_data_for_all_variants_tested_with_annotation.xlsx",dtype=dict(EXON=str,INTRON=str))
data = data.assign(CHROM=data.chrom.str.replace("chr", ""))
data = data.rename(columns = {
    "VCF_position" : "POSITION",
    "VCF_Ref" : "REF",
    "VCF_Alt" : "ALT"})
```

```python
subset = data[(data.Primary_consequence.isin({"missense_variant",'synonymous_variant'}))]
```

```python
subset[subset.Primary_consequence == 'synonymous_variant']
```

```python
from Bio.PDB.Polypeptide import protein_letters_3to1
protein_letters_1to3 = {v: k.title() for k, v in protein_letters_3to1.items()}
protein_letters_1to3['STOP'] = "Ter"
def make_hgvs_pro(r):
    aa_ref = protein_letters_1to3[r.ref_aa]
    aa_alt = protein_letters_1to3[r.alt_aa]
    if aa_ref == aa_alt:
        return f"p.{aa_ref}{r.Protein_position}="
    return f"p.{aa_ref}{r.Protein_position}{aa_alt}"
```

```python
subset = subset.assign(hgvs_pro=subset.apply(make_hgvs_pro,axis=1))
```

```python
subset = subset.rename(columns={"D15_combined_LFC":"score"})
```

```python
subset.to_csv(dataset_path / 'scoreset.csv',index=False)
```

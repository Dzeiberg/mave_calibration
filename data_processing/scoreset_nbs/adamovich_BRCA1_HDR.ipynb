```python
from pathlib import Path
import pandas as pd
import json
import mavehgvs
dataset_path = Path("/mnt/i/bio/mave_curation/adamovich_BRCA1_HDR/")
```

```python
data = pd.read_excel(dataset_path / "raw" / "mmc2.xlsx", sheet_name="Supp. Table S2")
```

```python
from Bio.PDB.Polypeptide import protein_letters_3to1
protein_letters_1to3 = {v:k.title() for k,v in protein_letters_3to1.items()}
def clean_variant_id(s):
    hgvs_pro = protein_letters_1to3[s[0]] + s[1:-1]
    if s[-1] == "*":
        hgvs_pro += "Ter"
    else:
        hgvs_pro += protein_letters_1to3[s[-1]]
    return "p."+ hgvs_pro
```

```python
data = data.assign(hgvs_pro=data.variantID.apply(clean_variant_id),).rename(columns={"FS_average_BRCA1_siRNA":"score"})
```

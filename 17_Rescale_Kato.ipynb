```python
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = joblib.load("/mnt/i/bio/mave_curation/Kato_TP53_AIP1nWT/observations.pkl")
```

```python
fig, ax = plt.subplots(1,3, figsize=(15,5),sharex=True,sharey=True)
sns.histplot(data['p_lp'],ax=ax[0],stat='density')
sns.histplot(data['b_lb'],ax=ax[1],stat='density')
sns.histplot(data['gnomad'],ax=ax[2],stat='density')
mu,sigma = data['b_lb'].mean(),data['b_lb'].std()
```

```python
fig, ax = plt.subplots(1,3, figsize=(15,5),sharex=True,sharey=True)
sns.histplot((data['p_lp'] - mu) / sigma,ax=ax[0],stat='density')
sns.histplot((data['b_lb'] - mu) / sigma,ax=ax[1],stat='density')
sns.histplot((data['gnomad'] - mu) / sigma,ax=ax[2],stat='density')
```
